```python
import numpy as np
import matplotlib.pyplot as plt
lam = np.arange(1e-9,3e-6,1e-9)
c , k , h = 3*1e8 , 1.3807e-23 , 6.626e-34
def plank(lam,T):
  a = 8*np.pi*h*c
  b = h*c/(lam*k*T)
  energydensity = (a/lam**5)*(1/(np.exp(b)-1))
  return energydensity
plt.plot(lam*1e10,plank(lam,4000),'r-')
plt.plot(lam*1e10,plank(lam,5000),'g-')
plt.plot(lam*1e10,plank(lam,6000),'b-')
plt.plot(lam*1e10,plank(lam,7000),'k-')
plt.axvline(4000,color='brown',ls='--')
plt.axvline(7500,color='brown',ls='--')
plt.ylabel('Intensity')
plt.xlabel('wavelenght')
plt.title('Blackbody Radiation',size='18')
plt.show()
```
